### Projeto - Análise das notificação casos de dengue no Brasil.
---
<p>O vírus da dengue é transmitido pela picada da fêmea do Aedes aegypti, um mosquito que costuma picar durante o dia (no início da manhã ou no final da tarde) e se multiplica em locais onde tem água parada. Ele vive dentro das casas e ao redor de residências, é nesse tipo de lugar que o mosquito encontra condições ideais para se proliferar como em quintais e calçadas, por isso os cuidados precisam ser contínuos.</p>

![Aedes aegypti](capa.png)

<p>Não existe um medicamento específico para tratar a dengue. Repousar e tomar muito líquido para evitar a desidratação são medidas importantes para aliviar os sintomas. Além disso, não é recomendado tomar remédios por conta própria. O uso de medicamentos compostos por ácido acetilsalicílico, como aspirina, Melhoral, AAS e de anti-inflamatórios como diclofenaco, nimesulida e ibuprofeno, podem aumentar o risco de sangramentos. </p>

##### Não se esqueça, casos de dengue ou suspeitos precisam de orientação médica.


### Sinais e Sintomas

- Dor abdominal intensa e contínua
- Vômitos persistentes
- Acúmulo de líquidos em cavidades corporais (ascite, derrame pleural, derrame pericárdico)
- Hipotensão postural e/ou lipotímia
- Letargia e/ou irritabilidade
- Aumento do tamanho do fígado (hepatomegalia) > 2cm
- Sangramento de mucosa

### **2. Processo de tratamento dos dados.**

Já obtido informações realizadas no tópico anterior, vamos agora trata-las.
O tratamento de dados ajuda a garantir que os dados estejam limpos e livres de erros, o que melhora a qualidade das análises.
Dados tratados resultam em análises mais precisas e confiáveis.

---
#### **2.1. Eliminando colunas desnecessárias**

```python
#Eliminando a coluna UF. Já temos uma coluna que referência o estado, sendo desnecessário duas.
df = df.drop('UF',axis=1)
display(df)
```

Ao importar essas base de dados, identifique duas colunas que referenciam ao mesmo sentido, ESTADO e UF. Dessa forma, a coluna UF é desnecessária, sendo assim escolhido a utilização da coluna ESTADO.

---
#### **2.2. Eliminando valores NaN**
São valores que não agregram em nada na nossa análise, além de ocupar espaço.

```python
#Elimninando valores NaN do df
df = df.dropna(how='any',axis=0,inplace=True)
```

---
#### **2.3. Eliminando valores duplicados**

```python
#Elimninando valores duplicados 
df = df.drop_duplicates(subset=['ID NOTIF'])
```

Anteriormente, as colunas ID NOTIF e IDADE apresentaram valores duplicados. A removação de valores duplicados deve ocorrer apenas na ID NOTIF, visto que a idade faz sentido ter valores repitidos, pessoas diferentes podem ter a mesma idade.

---
#### **2.4. Convertendo tipos de dados**

```python
#Convertendo a coluna DATA do tipo object para datetime
df['DATA'] = pd.to_datetime(df['DATA'], format="%d/%m/%Y")
display(df)
```

```python
#Convertendo a coluna IDADE do tipo Float para Int
df['IDADE'] = df['IDADE'].astype('int')
display(df)
```

Como dito acima, a coluna IDADE foi importada com o tipo de dado FLOAT, porém, não faz sentido que esse coluna mantenha esse tipo de dado, a idade é representada por um valor inteiro.

### **4. Análise gráfica**

Uma análise gráfica desempenha um papel fundamental na compreensão e comunicação de padrões, tendências e insights presentes nos dados

---
#### **4.1. Análise por região**

```python
#Total de notificação por região
noti_por_regi = df['REGIÃO'].value_counts()

#Criação da figura do gráfico
fig, ax = plt.subplots(figsize=(12, 6))
noti_por = ax.bar(noti_por_regi.index.tolist(), noti_por_regi.tolist(),color='Purple')

# Título
ax.text(-0.025, 1.18,'Notificação de casos de dengue por cada região do Brasil',
        fontsize=20,
        color='gray',
        fontweight='medium',
        transform=ax.transAxes
       )

# Subítulo
ax.text(-0.015, 1.06,'É percecptível a diferença de notificações entre a região Sudeste das demais.' 
'\nUm dos fatores que a ser considerado, é o fato da região abrigar as maiores capitais do Brasil. ',
        fontsize=12,
        color='gray',
        fontweight='medium',
        transform=ax.transAxes)

#Título do eixo X e Y
ax.set_xlabel('REGIÃO')

# Removendo a linha superior
ax.spines['top'].set_visible(False)

#Inserindo rótulos no gráfico
ax.bar_label(noti_por, fmt="%.01f", size=12,padding=-14,color="white")

#Deixando os rótulos do eixo X na diagonal
plt.xticks(rotation=45)

#Adicionando linhas ao gráfico
plt.grid(visible=True, axis='y')

#Configurando as labels
plt.yticks(range(0, max(noti_por_regi.tolist()) + 1, 40000))

#Apresentando o gráfico
plt.show()
```

- Por meio deste outro gráfico comprovamos que as maiores capitais do Brasil, presentes na região sudeste estão no tópicos 5 notificação de casos de dengue.

#### **4.4. Crescimento das notificações em um período de 30 dias.**

```python
#Agrupando a data e o sexo, e realizando uma conta de notificações por data.
data_noti = df.groupby(['DATA','SEXO'])['ID NOTIF'].count()
data_noti_df = data_noti.reset_index()

#Criando a figura do gráfico
plt.figure(figsize=(10, 4))

#Criando o gráfico de linha
sns.lineplot(data=data_noti_df,
             x='DATA',
             y='ID NOTIF',
             hue='SEXO',
             marker='o'
            )

#Deixando os rótulos do eixo X na diagonal
plt.xticks(rotation=45)

#Modificando os eixos
plt.xlabel('Data')
plt.title('Crescimento nas notificações no decorrer dos dias.')

#Adicionando linhas ao gráfico
plt.grid(visible=True, axis='y')

#Apresentando o gráfico
plt.show()
```

---
#### **4.5. Distribuição das notificações por gênero.**

Percebemos uma crescente conforme os dias. Entre as datas 21/01/2024 a 25/01/2024 percebemos que atingiu o ápice das notificações. Tanto o sexo Feminino quando o sexo masculino. 

```python
#Quantidade de notificações por sexo
noti_por_sexo = df.groupby('SEXO').size()

# Construindo a figura do gráfico
plt.figure(figsize=(6, 6))

# Gráfico de pizza
plt.pie(noti_por_sexo,
        labels=noti_por_sexo.index, 
        autopct='%1.1f%%', 
        startangle=90,
        pctdistance=1.25, 
        labeldistance=.6,
        shadow=True,
        colors=['purple', 'yellow', 'pink']
       )
#Legenda
plt.legend(loc='upper right')
#Título
plt.title('Distribuição de Notificações por Gênero')
# Equal aspect ratio para garantir que o gráfico seja um círculo
plt.axis('equal')  
plt.show()
```
